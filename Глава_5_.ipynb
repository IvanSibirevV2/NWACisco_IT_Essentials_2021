# Глава 5. Сети: основные понятия
Компьютерные сети позволяют пользователям совместно пользоваться ресурсами и обмениваться информацией. Можете ли вы представить мир без электронной почты, онлайн-газет, блогов, веб-сайтов и прочих сервисов, доступных через Интернет? Благодаря сетям пользователи могут также совместно пользоваться разными ресурсами, например, принтерами, приложениями, файлами, каталогами и дисковыми накопителями. В этой главе рассматриваются принципы и стандарты организации сетей, а также их назначение. ИТ-специалисты должны разбираться в основных сетевых понятиях, чтобы удовлетворять ожидания и потребности клиентов и пользователей сети.

Вы узнаете об основах проектирования сети и о том, как сетевые устройства влияют на передачу данных. К таким устройствам относятся концентраторы, коммутаторы, точки доступа, маршрутизаторы и межсетевые экраны. Здесь также рассматриваются различные типы подключений к сети Интернет, такие как DSL, кабельные линии, мобильный и спутниковый доступ. Вы узнаете о четырех уровнях модели TCP/IP, а также о функциях и протоколах, связанных с каждым из них. Вы также узнаете о различных беспроводных сетях и протоколах, в том числе о протоколах беспроводной сети LAN IEEE 802.11, о протоколах беспроводной связи на коротких расстояниях, включая RFID, NFC, и о стандартах протоколов для «умного дома», таких как Zigbee и Z-Wave. Эти знания помогут вам успешно проектировать, создавать и обслуживать сети. В конце главы мы обсудим различные типы сетевых кабелей: витую пару, волоконно-оптические и коаксиальные кабели. Вы узнаете о конструкции кабелей каждого типа, о способе передачи сигнала и о целесообразных вариантах их использования.

Важно получить не только теоретические знания о компьютерных сетях и их компонентов, но и приобрести практические навыки. При изучении этой главы вы сделаете и протестируете прямой кабель для сети Ethernet на базе неэкранированной витой пары (UTP).

# Сетевые значки
Сети представляют собой системы, образованные каналами. Компьютерные сети соединяют друг с другом различные устройства и пользователей. Для обозначения разных частей компьютерной сети используют различные значки.

## Узловые устройства
Сетевые устройства, знакомые большинству пользователей, называют оконечными или узловыми устройствами (рисунок 1). Их называют оконечными, поскольку такие устройства располагаются в конечных точках или на границе сети. Узловыми их называют из-за того, что они обычно на них установлены сетевые приложения, такие как веб-браузеры и почтовые клиенты, использующие сеть для предоставления различных услуг пользователям.


## Промежуточные устройства
Между узловыми устройствами в компьютерных сетях используется целый ряд других устройств. Такие промежуточные устройства обеспечивают обмен данными между узловыми устройствами. Наиболее распространенные промежуточные устройства показаны на рисунке 2.

- Коммутатор — подключает несколько устройств к сети.
- Маршрутизатор — пересылает трафик между сетями.
- Беспроводной маршрутизатор — подключает к сети несколько беспроводных устройств и может также содержать в себе коммутатор для подключения проводных устройств.
- Точка доступа (Access Point, AP) — подключается к беспроводному маршрутизатору и используется для расширения зоны покрытия беспроводной сети.
- Модем — подключает к сети Интернет домашнюю сеть или сеть небольшого офиса.


# Сетевые среды

Обмен данными по сети осуществляется в сетевой среде. Сетевая среда создает канал, по которому сообщение передается от источника к получателю. Существует несколько типов сетевых сред. На рисунке 3 показаны значки различных типов сетевых сред. Локальные вычислительные сети (LAN), глобальные сети (WAN) и беспроводные сети будут рассмотрены далее в этом разделе. Облако в сетевой топологии обычно обозначает подключения к Интернету. Интернет обычно служит средой для обмена данными между сетями.


# PAN
PAN (Personal Area Network) - это персональная сеть, соединяющая такие устройства, как мыши, клавиатуры, принтеры, смартфоны и планшеты, которые используются одним человеком. Чаще всего для подключения этих устройств используется технология Bluetooth. Bluetooth -это беспроводная технология, позволяющая устройствам взаимодействовать друг с другом на небольшом расстоянии.


# VLAN
Традиционно под локальной вычислительной сетью (LAN)

понимается сеть, которая соединяет устройства металлическими кабелями в пределах небольшой географической территории. В современном мире отличительной чертой локальной сети можно назвать ее принадлежность одному владельцу, например, домохозяйству или малому предприятию, или ее эксплуатацию одной ИТ-службой, например, в учебном заведении или в корпорации.

# VLAN
Виртуальные сети LAN (VLAN) позволяют администратору сегментировать порты на одном коммутаторе, как если бы он состоял из нескольких коммутаторов. Это повышает эффективность пересылки данных за счет того, что трафик ограничивается определенными портами. Кроме того, сети VLAN позволяют группировать оконечные устройства в целях их администрирования. На данной схеме VLAN 2 - это виртуальная локальная сеть для компьютеров ИТ-службы, даже если они расположены на разных этажах, при этом разрешения для этой сети могут отличаться от разрешений для других сетей VLAN.


# WLAN
Беспроводная сеть LAN (WLAN) аналогична сети LAN, однако в ней для подключения пользователей и устройств на небольшой географической территории вместо кабелей используется беспроводная связь. Передача данных в сети WLAN между беспроводными устройствами осуществляется с помощью радиосигналов.


-------------------

# WMN
Беспроводная ячеистая сеть (WMN) использует несколько точек доступа для расширения покрытия сети WLAN. В топологии присутствует беспроводной маршрутизатор. Две беспроводных точки доступа расширяют покрытие сети WLAN внутри дома. Коммерческие предприятия и муниципальные организации также могут использовать сети WMN для быстрого расширения зоны покрытия.


# MAN
Муниципальная сеть (MAN) - это сеть в пределах квартала или города. Сеть состоит из нескольких зданий, подключенных по беспроводной сети или по оптоволоконным линиям.


------------------

# WAN
Глобальная сеть (WAN) объединяет несколько сетей, рассредоточенных в различных географических регионах, Физические лица и организации получают доступ к WAN через операторов связи. Оператор связи обеспечивает доступ ваших стационарных или мобильных устройств к крупнейшей сети WAN, т. е. к сети Интернет. На рисунке показаны сети из Москвы и Токио, соединенные через Интернет.


# VPN
Виртуальная частная сеть (VPN) используется для защищенного подключения к другой сети через открытую сеть (например, Интернет). Чаще всего сети VPN используют для доступа удаленных работников к закрытой корпоративной сети. Удаленные работники - это пользователи сети, работающие за пределами объекта или удаленно. На рисунке жирной линией показано VPN-подключение удаленного работника 1 к маршрутизатору в центральном офисе компании. Удаленный работник 1 безопасно входит в сеть компании с помощью программного обеспечения VPN. Удаленный работник 2 не использует защищенное подключение и не сможет получить доступ к ресурсам компании.

# Краткая история технологий подключения
В 90-х годах скорости подключения к Интернету были ниже, чем сегодня. В наше время пропускная способность позволяет передавать голос, видео и данные. Для коммутируемого доступа нужен либо внутренний модем, установленный в компьютере, либо внешний модем, подключенный через USB. Линейный порт модема подключается к телефонной розетке с разъемом RJ-11. После физической установки модема его необходимо подключить к одному из программных COM-портов компьютера. Кроме того в модеме необходимо настроить параметры набора номера, например, код выхода на внешнюю телефонную линию и код города.

Мастер установки подключений или мастер сетевых подключений позволяет настроить подключение к серверу интернет-провайдера. Сегодня вместо телефонной линии для доступа в Интернет используется широкополосная связь:

Аналоговый телефон

При доступе к Интернету через аналоговый телефон данные передаются по стандартной телефонной линии. В этой схеме используется аналоговый модем, который «созванивается» с другим модемом, расположенным на удаленном объекте. Такой тип подключения называют коммутируемым доступом.

Цифровая сеть с интегрированными услугами

Цифровая сеть с интегрированными услугами (Integrated Services Digital Network, ISDN) использует несколько каналов и может предоставлять различные услуги. Поэтому ISDN относится к одному из видов широкополосной связи. Стандарт ISDN предполагает использование нескольких каналов для передачи голоса, видео и данных по обычным телефонным линиям. Пропускная способность ISDN больше, чем при традиционном коммутируемом доступе.

Широкополосный доступ

Широкополосная технология подразумевает использование различных частот для передачи нескольких сигналов в одной среде передачи. Например, по коаксиальному кабелю, который используется для доставки кабельного телевидения, можно наряду с сотнями ТВ-каналов передавать данные компьютерной сети. Ваш мобильный телефон может принимать голосовые вызовы и одновременно с этим использовать веб-браузер.

К наиболее распространенным типам широкополосных подключений можно отнести кабельную линию, цифровую абонентскую линию (Digital Subscriber Line, DSL), ISDN, спутниковые и мобильные подключения. На рисунке показано оборудование, используемое для подключения к широкополосной сети или широкополосной передачи данных.

# Цифровая абонентская линия (DSL), кабельная и оптоволоконная линии
Как в случае с DSL, так и в случае с кабельной линией для подключения к Интернету через интернет-провайдера используется модем, что показано на рисунке. DSL-модем подключает сеть пользователя непосредственно к цифровой инфраструктуре телефонной компании. Кабельный модем подключает сеть пользователя к оператору кабельной сети.

Цифровая абонентская линия (DSL)

DSL относится к непрерывно доступным службам, т. е. нет необходимости каждый раз набирать номер для подключения к Интернету. Голос и данные передаются по медному телефонному кабелю на разных частотах. Специальный фильтр предотвращает взаимные помехи между сигналами DSL и телефонными сигналами.

Кабельная линия

Для кабельного подключения к Интернету телефонная линия не нужна. В этом случае используется коаксиальный кабель, изначально предназначенный для передачи кабельного телевидения. Кабельный модем подключает ваш компьютер к оператору кабельной сети. Компьютер можно подключать непосредственно к кабельному модему. Однако если подключить к модему маршрутизатор, то через него к Интернету можно будет подключить несколько компьютеров.

Оптоволоконная линия

Оптоволоконные кабели изготавливаются из стекла или пластмассы, для передачи данных в них используется свет. У них очень высокая пропускная способность, позволяющая передавать большие объемы данных. Данные, передаваемые в рамках вашего подключения к Интернету, в определенный момент проходят через оптоволоконные сегменты сети. Оптоволоконные кабели используются в магистральных сетях, в крупных корпоративных сетях и в больших ЦОД. Устаревшая инфраструктура медных кабелей, окружающая домовладения и офисы, все чаще уступает место оптоволоконным кабелям. В качестве примера на рисунке показано кабельное подключение с использованием гибридной оптоволоконно-коаксиальной сети (Hybrid Fiber Coaxial, HFC), где на участке последней мили к дому пользователя используется оптоволоконный кабель. В самом доме сеть снова переходит на медный коаксиальный кабель.

Доступные варианты подключения зависят от географического положения и наличия операторов связи.
